```python
# importing basic libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing basic libraries
import numpy as np
import pandas as pd
```

```python
# Reading the dataset
job = pd.read_csv('jobs.csv')
clicks = pd.read_csv('job_clicks.csv')
```

```python
job.head()
```

```python
job = job.rename(columns = {'jobID' : 'jobId'})
```

```python
clicks.head()
```

```python
# Merging 2 dataframe on JobID
df = pd.merge(job , clicks , how = 'inner' , on = 'jobId')
```

```python
df.head()
```

```python
df[df['userId']==1]
```

```python
# Created pivot table which has userID as index and jobId as columns and has clicks as values
df = df.pivot_table(index = 'userId' , columns = 'jobId' , values = 'Clicks')
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.dropna(thresh = 10  , axis = 1)
```

```python
df.shape
```

```python
# Correlation matrix to check which jon is correlated to another job
job_df = df.corr()
```

```python
job_df.head()
```

```python
job_df.columns
```

```python
jobmat = job_df[1061]
```

```python
# Recommending jobs based on collabrative filtering
similar_it = job_df.corrwith(jobmat)
```

```python
similar_it = similar_it.sort_values(ascending = False)
```
